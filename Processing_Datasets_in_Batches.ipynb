The objective of this excercise is to write a user defined function which will process database data in chunks and add a new column which will be the product of two other columns from the database.

The database imported is the 'World Indicator Population Dataset'. Using the 2 columns: Total Population, Urban population (% of total), calculate the total urban population for countries with specific country codes.

```python
#importing required libraries
import pandas as pd
from google.colab import files
```

```python
#defining the new user-defined function
def db_processer(filename, country_code):
  """this function will take database names and a specific country code as arguments and process the database data in chunks"""

  #defining an empty database where each processed data will be added in each iteration
  altered_db = pd.DataFrame()

  #uploading files from database
  files.upload()

  #reading the data from a database in chunks
  original_db = pd.read_csv(filename, chunksize = 1000)

  #writing the for loop for checking condition and adding them to new database
  for chunk in original_db:
    required_country_data = chunk[chunk['CountryCode']== country_code]

    #combining the 2 columns of interest into a tuple and list for easier computation
    urban_population_tuple = zip(required_country_data['Total Population'], required_country_data['Urban population (% of total)'])
    urban_population_list = list(urban_population_tuple)

    #adding a new column to the required country database using list comprehensions:
    required_country_data['Total Urban Population'] = [int(tup[0]*tup[1]*0.01) for tup in urban_population_list]

    #adding the above information into the new database for each chunk
    altered_db = pd.concat([altered_db, required_country_data])

  return altered_db
```

```python
#calling the new function with the defined database with different country codes
result = db_processer('world_ind_pop_data.csv','CSS')
print(result)
```
